```python
import os
import cv2

def images_to_video(image_folder, output_video, fps=24):
    # 1. 获取文件夹内所有图片，按文件名自然排序
    img_ext = (".jpg", ".jpeg", ".png", ".bmp", ".tiff")
    file_list = [f for f in os.listdir(image_folder) if f.lower().endswith(img_ext)]
    # 按文件名升序（001.jpg, 002.jpg 顺序）
    file_list.sort()
    if not file_list:
        print("文件夹无图片！")
        return

    # 读取第一张图获取分辨率
    first_img_path = os.path.join(image_folder, file_list[0])
    frame = cv2.imread(first_img_path)
    h, w = frame.shape[:2]

    # 视频编码器 mp4
    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    writer = cv2.VideoWriter(output_video, fourcc, fps, (w, h))

    # 逐帧写入
    total = len(file_list)
    for idx, name in enumerate(file_list):
        img_path = os.path.join(image_folder, name)
        img = cv2.imread(img_path)
        if img is None:
            print(f"跳过损坏图片：{name}")
            continue
        writer.write(img)
        print(f"已处理 {idx+1}/{total}: {name}")

    writer.release()
    print(f"视频生成完成：{output_video}")

# ========== 修改这里参数 ==========
if __name__ == "__main__":
    IMAGE_DIR = r"predict_vis"       # 图片文件夹路径
    OUTPUT = r"./output.mp4"    # 输出视频名
    FPS = 20                   # 帧率
    images_to_video(IMAGE_DIR, OUTPUT, FPS)
```
